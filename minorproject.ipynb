# Spliting The Dataset 

```python
x=df.drop(columns='PlacementStatus')
y=df.PlacementStatus
x
```

```python
y
```

# Training The Dataset

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=42)
x_train
```

```python
x_test
```

```python
y_train
```

# Applying The Algorithms On Our Models

```python
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix,ConfusionMatrixDisplay
accuracies=[]
xgb_model=XGBClassifier(n_estimators=1000,learning_rate=0.05,objective="binary:logistic",eval_metric="logloss",random_state=70)
xgb_model.fit(x_train,y_train)
y_pred=xgb_model.predict(x_test)
accuracy_xgb=accuracy_score(y_test,y_pred)
print(f"Accuracy of XGB Model: {accuracy_xgb}")
cm=confusion_matrix(y_test,y_pred,normalize='true')
target_name=['Placed','Not Placed']
ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=target_name).plot(cmap='Blues')
plt.title('XGBoost Model')
```

```python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
lr.fit(x_train,y_train)
```

```python
prediction=lr.predict(x_test)
accuracy=accuracy_score(prediction,y_test)
print(f"Logistic Regression Accuracy: {accuracy}")
accuracies.append({'Model':'Logistic Regression','Accuracy':'accuracy'})
target_name=['Not Placed','Placed']
print(classification_report(y_test,prediction,target_names=target_name))
cm=confusion_matrix(y_test,prediction,normalize='true')
ConfusionMatrixDisplay(confusion_matrix=cm,display_labels=target_name).plot(cmap='Reds')
plt.title('Logistic Regression')

```

```python
from sklearn.tree import DecisionTreeClassifier
d_model=DecisionTreeClassifier(random_state=151)
d_model.fit(x_train,y_train)
y_pred_d=d_model.predict(x_test)
accuracy_d=accuracy_score(y_test,y_pred_d)
print(f"Decision Tree Accuracy:{accuracy_d}")
accuracies.append({'Model':'Decision Tree','Accuracy':accuracy_d})
cm_d=confusion_matrix(y_test,y_pred_d,normalize='true')
ConfusionMatrixDisplay(confusion_matrix=cm_d,display_labels=target_name).plot(cmap='Greens')
plt.title("Decision Tree")
```

```python
from sklearn.naive_bayes import GaussianNB
nb_model=GaussianNB()
nb_model.fit(x_train,y_train)
y_pred_nb=nb_model.predict(x_test)
accuracy_nb=accuracy_score(y_test,y_pred_nb)
print(f"Naive Bayes Accuracy:{accuracy_nb}")
accuracies.append({'Model':'Naive Bayes','Accuracy':accuracy_nb})
cm_nb=confusion_matrix(y_test,y_pred_nb,normalize='true')
ConfusionMatrixDisplay(confusion_matrix=cm_nb,display_labels=target_name).plot(cmap='Blues')
plt.title("Naive Bayes")
```

```python
from sklearn.svm import SVC
svm_model=SVC(kernel='linear',random_state=201)
svm_model.fit(x_train,y_train)
y_pred_svm=svm_model.predict(x_test)
accuracy_svm=accuracy_score(y_test,y_pred_svm)
print(f"SVM Accuracy:{accuracy_svm}")
accuracies.append({'Model':'Support Vector Machine','Accuracy':accuracy_svm})
cm_svm=confusion_matrix(y_test,y_pred_svm,normalize='true')
ConfusionMatrixDisplay(confusion_matrix=cm_svm,display_labels=target_name).plot(cmap='Reds')
plt.title("Support Vector Machine")
```

```python
from sklearn.ensemble import RandomForestClassifier
rfc_model=RandomForestClassifier(random_state=42)
rfc_model.fit(x_train,y_train)
y_pred_rfc=rfc_model.predict(x_test)
accuracy_rfc=accuracy_score(y_test,y_pred_rfc)
print(f"Random Forest Accuracy:{accuracy_rfc}")
accuracies.append({'Model':'Random Forest','Accuracy':accuracy_rfc})
cm_rfc=confusion_matrix(y_test,y_pred_rfc,normalize='true')
ConfusionMatrixDisplay(confusion_matrix=cm_rfc,display_labels=target_name).plot(cmap='Greens')
plt.title("Ramdom Forest")
```

# Importing Performance Matrics for Evaluation

```python
import matplotlib.pyplot as plt
from sklearn.metrics import ConfusionMatrixDisplay
from sklearn.metrics import precision_score, recall_score, f1_score, confusion_matrix, accuracy_score
models = {'XG-Boost': xgb_model, 'Logistic Regression': lr, 'Decision Trees': d_model, 'Naive Bayes': nb_model, 'SVM': svm_model,
         'Random Forest':rfc_model}

# Create a list to store accuracies
accuracies = []

# Iterate over models
for model_name, model in models.items():
    # Make predictions
    y_pred = model.predict(x_test)

    # Evaluate the model
    accuracy = accuracy_score(y_test, y_pred)


    # Precision, Recall, F1 Score, Specificity
    precision = precision_score(y_test, y_pred)
    recall = recall_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred)

    # Calculate specificity
    tn, fp, fn, tp = cm.ravel()
    specificity = tn / (tn + fp)

    # Append metrics to the accuracies list
    accuracies.append({
        'Model': model_name,
        'Accuracy': accuracy,
        'Precision': precision,
        'Recall': recall,
        'F1 Score': f1,
        'Specificity': specificity
    })
```

# Making Predictions

```python
from tabulate import tabulate
from termcolor import colored  # Add this import statement

accuracy_df = pd.DataFrame(accuracies)

# Remove duplicate rows based on the 'Model' column
accuracy_df = accuracy_df.drop_duplicates(subset='Model', keep='first')

# Display the accuracies in a beautified and colored tabular form
headers = accuracy_df.columns
data = accuracy_df.values  # Make sure 'data' is defined here

colored_table = [[colored(cell, 'blue', attrs=['bold']) if i == 0 else cell for i, cell in enumerate(row)] for row in data]
print(tabulate(colored_table, headers=headers, tablefmt='fancy_grid', showindex=False))

# Define the input features in the same order as the model was trained
feature_names = ['CGPA', 'Internships', 'Projects', 'Workshops/Certifications', 'AptitudeTestScore',
                 'SoftSkillsRating', 'ExtracurricularActivities', 'PlacementTraining', 'SSC_Marks', 'HSC_Marks']
print("ExtracurricularActivities and PlacementTraining should be in form of 1(yes)or 0(no)")
# Function to get input from the user
def get_user_input():
    user_input = []
    
    for feature in feature_names:
        value = float(input(f"Enter {feature}: "))
        user_input.append(value)
    return np.array(user_input).reshape(1, -1)

# Get user input
user_data = get_user_input()

# Make prediction
placement_prediction = nb_model.predict(user_data)

# Print the prediction
if placement_prediction[0] == 1:
    print("The Naive Bayes model predicts that the student will be placed.")
else:
    print("The Naive Bayes model predicts that the student will not be placed.")
```
